```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

studData = pd.read_csv('nas-pupil-marks.csv')
```

```python
######################       Q2       #######################

#WE are separating on the basis of gender and displaying the data of all the subjects 
#on the basis of difference between these averages, grouped by individual states.
#values greater than 0 indicates Boys performed better in that subject
#values greater than 1 indicates Girls performed better in that subject

#Program
boyData = studData[studData['Gender'] == 1].groupby('State').agg(
    {'Maths %': np.mean, 'Science %': np.mean, 'Reading %': np.mean, 'Social %': np.mean})
girlData = studData[studData['Gender'] == 2].groupby('State').agg(
    {'Maths %': np.mean, 'Science %': np.mean, 'Reading %': np.mean, 'Social %': np.mean})

q2 = boyData - girlData

q2.plot.bar()
plt.show()
```

```python
######################       Q3       #######################

#WE are separating on the basis of South and North Indian States and 
#displaying the data on the basis of difference between these averages, Separately for Science and Maths subject.
#values greater than 0 indicates North Indian States combined together perform better in than their counterpart
#values greater than 1 indicates South Indian States combined together perform better in that their counterpart

#Program
studData['State'] = studData['State'].astype(str)

southStates = studData[
    (studData['State'] == 'AP') | (studData['State'] == 'TN') | (studData['State'] == 'KA') | (studData['State'] == 'KL')
    | (studData['State'] == 'PY') | (studData['State'] == 'LD') | (studData['State'] == 'AN')]

northStates = studData[
    (studData['State'] != 'AP') | (studData['State'] != 'TN') | (studData['State'] != 'KA') | (studData['State'] != 'KL')
    | (studData['State'] != 'PY') | (studData['State'] != 'LD') | (studData['State'] != 'AN')]


southavg = southStates.agg({'Maths %': np.mean, 'Science %': np.mean})
northavg = northStates.agg({'Maths %': np.mean, 'Science %': np.mean})

q3 = southavg - northavg

q3.plot.bar()
plt.show()
```
